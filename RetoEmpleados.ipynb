```python
# 1) Importar las librer√≠as requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, OrdinalEncoder
from sklearn.decomposition import PCA

# Opcional: configuraci√≥n de visualizaci√≥n
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 160)
```

```python
# 2) Leer el CSV y colocar los datos en un DataFrame llamado EmpleadosAttrition
ruta_csv = 'empleadosRETO.csv'  # Ajustar la ruta si es necesario
EmpleadosAttrition = pd.read_csv(ruta_csv)
EmpleadosAttrition.shape, EmpleadosAttrition.head(3)
```

```python
# 3) Eliminar columnas sin relaci√≥n con la salida
cols_a_eliminar = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']
cols_existentes = [c for c in cols_a_eliminar if c in EmpleadosAttrition.columns]
EmpleadosAttrition = EmpleadosAttrition.drop(columns=cols_existentes)
EmpleadosAttrition.columns
```

```python
# 4) 5) y 6) Homogeneizar HiringDate y crear Year y YearsAtCompany
# Nota: si HiringDate viene como texto, se parsea a datetime; si ya es datetime, se respeta
if 'HiringDate' in EmpleadosAttrition.columns:
    EmpleadosAttrition['HiringDate'] = pd.to_datetime(EmpleadosAttrition['HiringDate'], errors='coerce', infer_datetime_format=True)
    # Year como entero
    EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].dt.year.astype('Int64')
    # YearsAtCompany = 2018 - Year
    EmpleadosAttrition['YearsAtCompany'] = (2018 - EmpleadosAttrition['Year']).astype('Int64')
else:
    # Si no existe HiringDate, se crea YearsAtCompany como NaN (o se omite)
    EmpleadosAttrition['Year'] = pd.Series(dtype='Int64')
    EmpleadosAttrition['YearsAtCompany'] = pd.Series(dtype='Int64')

EmpleadosAttrition[['HiringDate','Year','YearsAtCompany']].head(5)
```

```python
# 7) 8) 9) Limpieza de DistanceFromHome: renombrar y crear versi√≥n entera sin 'km'
if 'DistanceFromHome' in EmpleadosAttrition.columns:
    EmpleadosAttrition = EmpleadosAttrition.rename(columns={'DistanceFromHome': 'DistanceFromHome_km'})
    # Extraer n√∫meros de strings como '12 km', '12km', '12 KM', etc.
    EmpleadosAttrition['DistanceFromHome'] = (
        EmpleadosAttrition['DistanceFromHome_km']
        .astype(str)
        .str.extract(r'(\d+(?:\.\d+)?)', expand=False)
        .astype(float)
        .round()
        .astype('Int64')
    )

EmpleadosAttrition[['DistanceFromHome_km','DistanceFromHome']].head(5)
```

Los valores faltantes que fueron detectados durante la exploraci√≥n preliminar son del tipo categ√≥rico (ordinal) por lo que pueden imputados a trav√©s de la moda de cada caracter√≠stica. No se seleccionaron las alternativas de:

A fin de satisfacer la suposici√≥n de normalidad, se realiza una prueba de Shapiro-Wilk para evaluar la distribuci√≥n de las variables cuantitativas:

```python
# Extra Opcional 2) Verificar
from scipy.stats import shapiro

num_cols = EmpleadosAttrition.select_dtypes(include=[np.number]).columns.tolist()

normality_results = {}

for col in num_cols:
    s = EmpleadosAttrition[col].dropna()

    # Shapiro requiere al menos 3 valores
    if s.size >= 3:
        stat, p_value = shapiro(s.sample(min(len(s), 400), random_state=42))
        normality_results[col] = {
            'statistic': stat,
            'p_value': p_value,
            'is_normal_Œ±0.05': p_value > 0.05
        }

normality_results
```

De la evaluaci√≥n anterior se concluye que ninguna de las variables cuantitativas sigue una distribuci√≥n normal, por lo que

```python
# Extra Opcional 3) B√∫squeda r√°pida de valores at√≠picos (+/- 3 sigma) en variables num√©ricas (informativo, no elimina)
num_cols = EmpleadosAttrition.select_dtypes(include=[np.number]).columns.tolist()
outliers_info = {}
for col in num_cols:
    s = EmpleadosAttrition[col].dropna()
    if s.size > 0:
        mu, sigma = s.mean(), s.std(ddof=1)
        lower, upper = mu - 3*sigma, mu + 3*sigma
        mask_out = (s < lower) | (s > upper)
        outliers_info[col] = int(mask_out.sum())
outliers_info  # en el dictamen se especifican los valores at√≠picos por columna
```

 Con base en los resultados de la prueba de distribuci√≥n normal, se puede concluir que, los valores identificados a m√°s de 3 desviaciones est√°ndar no son estrictamente at√≠picos.

**Nota:** el One-Hot encoding es un m√©todo sirve para transformar variables categ√≥ricas en variables dicot√≥micas. Cada uno de los valores de la variable categ√≥rica se transforma en una variable binaria (0, 1), por lo que aumenta el n√∫mero de caracter√≠sticas de la matriz a evaluar (n).

Dado que esto eleva el coste computacional y no sabemos si realmente estas nuevas variables aportan valor al an√°lisis, sugerir√≠a llevar a cabo sendas pruebas de Chi-cuadrado de independencia para evaluar las siguientes caracter√≠sticas: 'BusinessTravel', 'Department', 'OverTime', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus' contra 'Attrition'.

El siguiente paso es un extra opcional para evaluar el efecto de cada una de estas variables categ√≥ricas sobre 'Attrition':


```python
# Extra Opcional 5) Evaluaci√≥n de efecto de variables categ√≥ricas de entrada por prueba de Chi-cuadrada
from scipy.stats import chi2_contingency
categoricas = ['BusinessTravel', 'Department', 'OverTime', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Attrition']
categoricas_presentes = [c for c in categoricas if c in EmpleadosAttrition.columns]
chi2_results = []

for col in categoricas_presentes:
    if col == 'Attrition':
        continue  # no se compara consigo misma

    # Tabla de contingencia
    contingency = pd.crosstab(
        EmpleadosAttrition[col],
        EmpleadosAttrition['Attrition'],
        dropna=False
    )

    # Requisito m√≠nimo para chi-cuadrado
    if contingency.shape[0] > 1 and contingency.shape[1] > 1:
        chi2, p_value, dof, expected = chi2_contingency(contingency)

        chi2_results.append({
            'Variable': col,
            'Chi2': chi2,
            'p_value': p_value,
            'DoF': dof,
            'Significativa_Œ±0.05': p_value < 0.05
        })

chi2_df = pd.DataFrame(chi2_results).sort_values('p_value')

chi2_df
```

Del paso opcional anterior se obtiene que s√≥lo en:
'OverTime',
'JobRole', y
'MaritalStatus'
a un Œ± = 0.05, se rechaza la H‚Çí de que estos NO tienen correlaci√≥n con la variable 'Attrition'.

Por lo anterior, y con base en una extensi√≥n del paso 3 correr√© el an√°lisis sin las dem√°s variables categ√≥ricas ('BusinessTravel', 'Department', 'EducationField', 'Gender').

```python
# 14.1) Eliminar dummies dicot√≥micas que no aportan valor
cols_a_quitar = ['OverTime_No', 'Attrition_No']
cols_a_quitar = [c for c in cols_a_quitar if c in EmpleadosAttrition.columns]
if cols_a_quitar:
    EmpleadosAttrition = EmpleadosAttrition.drop(columns=cols_a_quitar)
```

```python
# 14.2) Calcular correlaci√≥n lineal de cada variable con Attrition
# La columna de salida despu√©s de dummies suele quedar como Attrition_Yes/Attrition_No; necesitamos identificar la positiva
# Buscamos la columna que represente attrition (si es Yes/No). Suponemos 'Attrition_Yes' como salida
attrition_cols = [c for c in EmpleadosAttrition.columns if c.startswith('Attrition_')]
if len(attrition_cols) == 0:
    raise ValueError("No se encontr√≥ columna de Attrition codificada. Verifica datos originales.")
# Elegimos 'Attrition_Yes' si existe; si no, usamos la primera
attrition_target = 'Attrition_Yes' if 'Attrition_Yes' in attrition_cols else attrition_cols[0]

# Correlaciones de Pearson contra la salida
corrs = EmpleadosAttrition.corr(numeric_only=True)[attrition_target].drop(labels=[attrition_target], errors='ignore')
corrs_sorted = corrs.sort_values(ascending=False)
corrs_sorted.head(10)
```

```python
# 15) Seleccionar variables con correlaci√≥n >= 0.1 y mantener la salida
umbral = 0.1
vars_seleccionadas = corrs.index[corrs.abs() >= umbral].tolist()

# Asegura que la columna de salida se incluya
cols_finales = vars_seleccionadas + [attrition_target]
cols_finales = [c for c in cols_finales if c in EmpleadosAttrition.columns]

EmpleadosAttritionFinal = EmpleadosAttrition[cols_finales].copy()
EmpleadosAttritionFinal.shape, EmpleadosAttritionFinal.head(3)
```

```python
# 16) PCA sobre EmpleadosAttritionFinal (sin la columna de salida)
# Convertir todas las columnas a tipo float y llenar cualquier valor NaN o pd.NA restante.
# Esto asegura que PCA reciba una matriz num√©rica limpia.
X_pca_input = EmpleadosAttritionFinal.drop(columns=[attrition_target]).astype(float)

# Llenar cualquier NaN restante, por ejemplo, con la media de la columna.
# Dada la preparaci√≥n previa, los NaNs deber√≠an ser m√≠nimos, pero es una medida de seguridad.
X_pca_input = X_pca_input.fillna(X_pca_input.mean())

X = X_pca_input.values
pca = PCA(n_components=None, svd_solver='full', random_state=42)
EmpleadosAttritionPCA = pca.fit_transform(X)

# Varianza explicada individual y acumulada
var_exp = pca.explained_variance_ratio_
varianza_acumulada = np.cumsum(var_exp)

#Scree plot para graficar la varianza
# √çndices de componentes
componentes = np.arange(1, len(var_exp) + 1)

# Gr√°fica
plt.figure()
plt.plot(componentes, var_exp, marker='o', label='Varianza explicada')
plt.plot(componentes, varianza_acumulada, marker='s', linestyle='--', label='Varianza acumulada')
plt.axhline(y=0.80, linestyle=':', label='80% varianza')
plt.xticks(componentes)
plt.xlabel('Componente principal')
plt.ylabel('Proporci√≥n de varianza explicada')
plt.title('Scree Plot ‚Äì PCA EmpleadosAttrition')
plt.legend()
plt.show()
```

```python
# 17) Agregar el m√≠nimo n√∫mero de componentes que expliquen al menos 80% de la varianza

# Dado que en las instrucciones no se espec√≠fica si esto debe dejarse programado como c√°lculo para permitir distintos an√°lisis o:
# puede tomarse como una constante del scree plot anterior, he decidido dejarlo as√≠, por un c√≥digo m√°s simple.

# N√∫mero de componentes necesarios
n_componentes_80 = 8   # seg√∫n el scree plot anterior

# Agregar los componentes al DataFrame
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(
    **{
        f'C{i}': EmpleadosAttritionPCA[:, i]
        for i in range(n_componentes_80 + 1)
    }
)

# Verificaci√≥n r√°pida
EmpleadosAttritionFinal.head()
```

```python
# 18) Guardar el dataset final en CSV

# Lista de columnas sin 'Attrition_Yes'
cols = [c for c in EmpleadosAttritionFinal.columns if c != 'Attrition_Yes']

# Reordenar: todas las dem√°s columnas + 'Attrition_Yes' al final
EmpleadosAttritionFinal = EmpleadosAttritionFinal[cols + ['Attrition_Yes']]

EmpleadosAttritionFinal.to_csv('EmpleadosAttritionFinal.csv', index=False)
```
